```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 17, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Alexandria'],
    'Salary': [100, 150, 200, 300]
}

df = pd.DataFrame(data)
df
```

```python
df[['Name', 'Salary']]
```

```python
df.iloc[1]
```

```python
df.describe()
# to convert it to rows not columns
df.describe().T
```

```python
print('The minimum Salary is ', df['Salary'].min())
print('The maximum Salary is ', df['Salary'].max())
print('The average Salary is ', df['Salary'].mean())
```
